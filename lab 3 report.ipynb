A resistor with zero value creates a short circuit, which can damage the circuit.  
Students need to test the resistance before use.  
Check if the resistance value is zero. If yes, print "Short Circuit", else print "Safe".
Store resistance in a variable, use if-else.


```python
resistance=float(input("please enter the resistance"))  #asking for resistance value from user.
if resistance==0:  
 print("short circuit")
else:
 print("safe")
 print("resistance",resistance)
```
