### Find Total Impedance 
---
1. Add the three impedances. 
2.	Print real part (resistive) and imaginary part (reactive). 
3. Print magnitude and phase angle (use abs() and cmath.phase()). 


```python
import cmath #importing modules
import math 
f=50
r=100
l=5
c=7
w=2*math.pi
z_r=complex(r,0)
z_l=complex(0,w*l)
z_c=complex(0,-1/w*c)
print(f"resistive impedence is: {z_r} ,inductive impedence is: {z_l} ,and capacitive impedence is:{z_c}")
#calculating sum of three impedence
z_total=z_r+z_l+z_c  
print("total impedence is :",z_total)
#printing real and imaginary parts
print("real part of total impedence: ",z_total.real)
print("imaginary part of total impedence: ",(z_total.imag))
#printing magnitude of total impedence   
print("magnitude of total impedence is :",abs(z_total)) 
phase_rad=cmath.phase(z_total)
phase_deg=phase_rad*(180/math.pi)
#printing phase angle in degree
print(f"phase angle of total impedence is:{phase_deg:.2f} degree")  


```
