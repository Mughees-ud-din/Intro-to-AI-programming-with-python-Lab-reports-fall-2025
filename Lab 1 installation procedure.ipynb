# This lab tells us how to install everything required for our futher learning

1. Anaconda = big bundle of data-science packages (easy for students). Download page: Anaconda Distribution.
[https://www.anaconda.com/](https://www.anaconda.com/) 
2. Windows: run the .exe installer, choose Just Me and do NOT check “Add Anaconda to my PATH” (Anaconda recommends using Anaconda Prompt instead).
3. Download & install Visual Studio Code from the official site (then open it)
[https://code.visualstudio.com/](https://code.visualstudio.com/) 
4. create env (pick a python version your course uses; e.g. 3.11)
conda create --name mycourse python=3.11 -y

4. Activate it
conda activate mycourse

5. Install typical packages
conda install numpy pandas matplotlib jupyter -y
6. OR use conda-forge for more packages:
conda install -c conda-forge scikit-learn seaborn -y
7.  conda install ipykernel -y
